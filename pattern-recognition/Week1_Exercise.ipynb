```python
# Copy the Species to new 'Species_label' column
iris_df['Species_label'] = iris_df['Species']
print(iris_df)
```

```python
Species = iris_df['Species_label'].unique() # Unique() return numpy array, not normal serires
for i in range(len(Species)):
  iris_df['Species_label'] = iris_df['Species_label'].replace(to_replace=Species[i],value=i)
iris_df['Species_label']
```

```python
display(iris_df.mean(numeric_only=True))
display(iris_df.median(numeric_only=True))
display(iris_df.min(numeric_only=True))
display(iris_df.max(numeric_only=True))
#display(iris_df.avg(numeric_only=True))
```

```python
def print_format(df,stat):
  print(f"{'-'*40} {stat} {'-'*40}\n")  # fixed number of dashes
  func = getattr(df.groupby('Species'), stat)
  display(func(numeric_only=True))
  print('\n')

print_format(iris_df, 'mean')
print_format(iris_df, 'var')
print_format(iris_df, 'min')
print_format(iris_df, 'max')
```

```python
#Visualization plitting scatter plot
import matplotlib.pyplot as plt
df0 = iris_df[iris_df['Species_label']==0]
x = df0['SepalLengthCm']
y = df0['SepalWidthCm']
plt.scatter(x, y, label=df0['Species'].unique()[0])
df1 = iris_df[iris_df['Species_label']==1]
x = df1['SepalLengthCm']
y = df1['SepalWidthCm']
plt.scatter(x, y, label=df1['Species'].unique()[0])
df2 = iris_df[iris_df['Species_label']==2]
x = df2['SepalLengthCm']
y = df2['SepalWidthCm']
plt.scatter(x, y, label=df2['Species'].unique()[0])

plt.xlabel('Sepal Length (Cm)')
plt.ylabel('Sepal Width (Cm)')
plt.title('Scatter plot for Species 0')
plt.legend()
plt.show()
```

```python
df0 = iris_df[iris_df['Species_label']==0]
x = df0['PetalLengthCm']
y = df0['PetalWidthCm']
plt.scatter(x, y, label=df0['Species'].unique()[0])
df1 = iris_df[iris_df['Species_label']==1]
x = df1['PetalLengthCm']
y = df1['PetalWidthCm']
plt.scatter(x, y, label=df1['Species'].unique()[0])
df2 = iris_df[iris_df['Species_label']==2]
x = df2['PetalLengthCm']
y = df2['PetalWidthCm']
plt.scatter(x, y, label=df2['Species'].unique()[0])

plt.xlabel('Petal Length (Cm)')
plt.ylabel('Petal Width (Cm)')
plt.title('Scatter plot for Species 0')
plt.legend()
plt.show()
```
